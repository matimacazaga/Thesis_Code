```python
import pickle
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request as url
import os.path
import numpy as np
from IPython.display import display
```

```python
#Cargo lista con tickers de acciones a utilizar
tickers = pickle.load(open('/home/matimacazaga/Documents/Universidad/Tesis/tickers.p', 'rb'))
```

```python
def getstock(ticker, update = False):
    """
    Descarga .csv con precios de cierre de la acción, volumen, etc o lee .csv ya existente
    #Input
    ticker : ticker de la acción
    update : si se desea actualizar el .csv 
    #Return
    PandasDataframe conteniendo los datos del archivo .csv
    """
    file_path = '/home/matimacazaga/Documents/Universidad/Tesis/CSV/'+ticker+'.csv'
    if os.path.exists(file_path) == False or update == True:
        url.urlretrieve('http://www.ravaonline.com/v2/empresas/precioshistoricos.php?e='+ticker+'&csv=1', file_path)
        return pd.read_csv(file_path,sep = ',')
    else:
        return pd.read_csv(file_path,sep = ',')    
```

```python
#Creo un DataFrame donde almacenar los valores del test
JB_values = pd.DataFrame(index = tickers, columns = ['JB','p-value'])
display(JB_values)
```

```python
no_usadas = []
for t in tickers:
    stock = getstock(t,update = False)
    stock = stock[stock['cierre'] != 0]
    if len(stock) >= 2000:
        returns = (stock['cierre'] - stock['cierre'].shift(1))/(stock['cierre'].shift(1))
        returns.dropna(inplace = True)
        jb_pvalue = stats.jarque_bera(returns.values)
        JB_values.loc[t,'JB']= jb_pvalue[0]
        JB_values.loc[t,'p-value'] = jb_pvalue[1]
    else:
        no_usadas.append(t)
```
