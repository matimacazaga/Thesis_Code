```python
from fbm import FBM
```

```python
f1 = FBM(n = 50, hurst = 0.85, length = 1, method = 'daviesharte')
f2 = FBM(n = 50, hurst = 0.25, length = 1, method = 'daviesharte')
f3 = FBM(n = 50, hurst = 0.5, length = 1, method = 'daviesharte')
```

```python
fbm_tendency = f1.fbm()
fbm_mean_reversion = f2.fbm()
fbm_brownian = f3.fbm()
```

```python
t_values = f1.times()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl 

%matplotlib inline

plt.style.use('seaborn-paper')
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
params = {
   'axes.labelsize': 10,
   'font.size': 10,
   'legend.fontsize': 10,
   'xtick.labelsize': 10,
   'ytick.labelsize': 10,
   'figure.figsize': [6, 4.5]
   }
mpl.rcParams.update(params)

plt.plot(t_values, fbm_tendency, linewidth=2, color='#1c6bf3',label = r'\textbf{H = 0.85}')
plt.plot(t_values, fbm_mean_reversion, linewidth=2, color='#ff4040',label = r'\textbf{H = 0.25}')
plt.plot(t_values, fbm_brownian, linewidth=2, color='#008000',label = r'\textbf{H = 0.5}')

plt.xlabel(r'\textbf{Tiempo}')
plt.ylabel(r'\textbf{$\displaystyle W_{H}(t)$}')
plt.legend()
plt.savefig('FBMexample.eps', format = 'eps', dpi = 1000)
```
