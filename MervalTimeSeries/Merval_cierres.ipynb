```python
import matplotlib.pyplot as plt 
import matplotlib as mpl 
import numpy as np 
from scipy.stats import norm
import urllib.request as url
import pandas as pd 
```

```python
file_path = '/home/matimacazaga/Documents/Universidad/Tesis/Informe/PythonCode/MervalTimeSeries/MervalPrices.csv'
#url.urlretrieve('http://www.ravaonline.com/v2/empresas/precioshistoricos.php?e=MERVAL&csv=1', file_path)
merval = pd.read_csv(file_path, sep = ',')
closes = merval['cierre']
dates = pd.to_datetime(merval['fecha'], yearfirst = True)
```

```python
%matplotlib inline
closes.dropna(inplace=True)

plt.style.use('seaborn-paper')
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
params = {
   'axes.labelsize': 10,
   'font.size': 10,
   'legend.fontsize': 10,
   'xtick.labelsize': 10,
   'ytick.labelsize': 10,
   'figure.figsize': [6, 4.5]
   }
mpl.rcParams.update(params)
plt.plot(dates, closes, linewidth=2, color='#1c6bf3',label =r"\textbf{Cierre diario Indice Merval}")

plt.xlabel(r'\textbf{Fecha}')
plt.ylabel(r'\textbf{Precio}')
plt.legend()
plt.savefig('closes.eps', format = 'eps', dpi = 1000)
```
